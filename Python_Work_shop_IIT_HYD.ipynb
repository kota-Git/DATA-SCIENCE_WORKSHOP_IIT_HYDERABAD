Inheritance means Acquiring the properties from the parent class (base class)

Base class(parent class) <---------------------derived class(child class)

1) variables -----------------------------------1)variables

2) methods-------------------------------------2)Methods

we create object for derived class and we can acces base class

```python
# signle inheritance ---means one base and once derived class i.e the object created for  derived  class and it can acces base class also


# Python code to demonstrate how parent constructors
# are called.

# parent/base  class
class Person():

	# __init__ is known as the constructor
	def __init__(self, name, idnumber):
		self.name = name
		self.idnumber = idnumber

	def display(self):
		print(self.name)
		print(self.idnumber)

	def details1(self):
		print("My name is ",self.name)
		print("IdNumber:",self.idnumber)

# child class/ derived class
class Employee(Person):
	def __init__(self, name, idnumber, salary, post):
		self.salary = salary
		self.post = post

		# invoking the __init__ of the parent class/ base class
		Person.__init__(self, name, idnumber)

	def details(self):
		print("My name is ",self.name)
		print("IdNumber:",self.idnumber)
		print("Post: ",self.post)


# creation of an object variable or an instance
a = Employee('laxman', 886012, 200000, "Intern")

# calling a function of the class Person using
# its instance
a.display()
a.details()
```

```python
# Multi-level inheritance means acquring the properties from parent class/ base class level wise

# example grand parent(base class) ----->parent(derived+base class)------>child



# A Python program to demonstrate inheritance

# Base or Super class. Note object in bracket.
# (Generally, object is made ancestor of all classes)
# In Python 3.x "class Person" is
# equivalent to "class Person(object)"
class Person(object):

	# Constructor
	def __init__(self, name):
		self.name = name

	# To get name
	def getName(self):
		return self.name

	# To check if this person is employee
	def isEmployee(self):
		return False


# Inherited or Sub class (Note Person in bracket)
class Employee(Person):

	# Here we return true
	def isEmployee(self):
		return True








# Driver code
emp = Person("ravik1") # An Object of Person
print(emp.getName(), emp.isEmployee())

emp = Employee("ravi2") # An Object of Employee
print(emp.getName(), emp.isEmployee())


```

```python
 # Hierarchical inheritance ---one parent has mulitiple childs

class parent():
  def pdisplay(self):
    print("PARENT")

class child1(parent):
  def c1display(self):
    print("CHILD-1")

class child2(parent):
  def c2display(self):
    print("CHILD-2")


c1=child1()
c1.pdisplay()
c1.c1display()
c2=child2()
c2.pdisplay()
c2.c2display()
```

```python

# mulitiple inheritance---two parents has one child

class f():
    def __init__(self, id):
        self.id =  id

    def details(self):
        print("this is id number", self.id)

class m():
    def __init__(self, name):
        self.name = name

    def show(self):
        print("this is the name", self.name)

class child(f, m):
    def __init__(self, id, name, salary):
        f.__init__(self, id)  # Call the constructor of the parent class 'f'
        m.__init__(self, name)  # Call the constructor of the parent class 'm'
        self.salary = salary

    def employee_details(self):
        print("this is id number", self.id)
        print("this is name", self.name)
        print("this is salary", self.salary)

obj = child(222, "Ravi", 1000)
obj.employee_details()
```

```python
# multilevel inheritance

class Grandparent:
    def __init__(self, grandparent_name):
        self.grandparent_name = grandparent_name

    def display_grandparent(self):
        print("Grandparent:", self.grandparent_name)

class Parent(Grandparent):
    def __init__(self, grandparent_name, parent_name):
        super().__init__(grandparent_name)
        self.parent_name = parent_name

    def display_parent(self):
        print("Parent:", self.parent_name)

class Child(Parent):
    def __init__(self, grandparent_name, parent_name, child_name):
        super().__init__(grandparent_name, parent_name)
        self.child_name = child_name

    def display_child(self):
        print("Child:", self.child_name)

# Create an instance of the Child class
child_obj = Child("Grandpa", "Dad", "John")

# Call methods to display information
child_obj.display_grandparent()
child_obj.display_parent()
child_obj.display_child()
```

```python
# hierachical inheritance

class Animal:
    def __init__(self, name, species, color):
        self.name = name
        self.species = species
        self.color = color

    def speak(self):
        pass  # This will be overridden by subclasses

class Cat(Animal):
    def __init__(self, name, species, color, breed):
        super().__init__(name, species, color)
        self.breed = breed

    def speak(self):
        return f"{self.name} the {self.color} {self.breed} cat says Meow!"

class Dog(Animal):
    def __init__(self, name, species, color, size):
        super().__init__(name, species, color)
        self.size = size

    def speak(self):
        return f"{self.name} the {self.color} {self.size} dog says Woof!"

# Create instances of Cat and Dog
cat = Cat("Whiskers", "Feline", "Gray", "Siamese")
dog = Dog("Fido", "Canine", "Brown", "Medium")

# Access instance variables and call the speak method on each instance
print(f"{cat.name} is a {cat.color} {cat.breed} cat.")
print(f"{dog.name} is a {dog.color} {dog.size} dog.")
print(cat.speak())  # Output: Whiskers the Gray Siamese cat says Meow!
print(dog.speak())  # Output: Fido the Brown Medium dog says Woof!
```
